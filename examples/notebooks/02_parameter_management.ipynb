# MetaPulsar Parameter Management Tutorial

This notebook demonstrates how to use the `ParameterManager` for astrophysical consistency across multiple PTA collaborations.

## Learning Objectives
- Understand parameter management concepts
- Learn how to create parameter mappings
- Explore different parameter combination strategies
- Handle parameter inconsistencies across PTAs


```python
from metapulsar import FileDiscoveryService, ParameterManager

print("MetaPulsar ParameterManager imported successfully!")
```

## 1. Discover Files and Prepare Data

First, let's discover files and prepare them for parameter management.


```python
# Create file discovery service
discovery = FileDiscoveryService()

# Discover files for specific PTAs
pta_names = ["epta_dr2", "ppta_dr2"]
file_data = discovery.discover_all_files_in_ptas(pta_names)

print("File discovery results:")
for pta_name, files in file_data.items():
    print(f"  {pta_name}: {len(files)} file pairs")

# Convert to single file per PTA format for ParameterManager
single_file_data = {}
for pta_name, file_list in file_data.items():
    if file_list:
        single_file_data[pta_name] = file_list[0]  # Take first file

print("\nSingle file data for ParameterManager:")
for pta_name, file_pair in single_file_data.items():
    print(f"  {pta_name}: {file_pair['par'].name} / {file_pair['tim'].name}")

if not single_file_data:
    print("\nNo file data available - this might happen in test environments")
    print("In a real environment, you would have actual par/tim files")
```

## 2. Create ParameterManager

The `ParameterManager` handles parameter consistency across multiple PTAs.


```python
if single_file_data:
    # Create ParameterManager
    print("Creating ParameterManager...")
    param_manager = ParameterManager(
        file_data=single_file_data,
        reference_pta="epta_dr2",
        combine_components=["astrometry", "spindown", "binary", "dispersion"],
        add_dm_derivatives=True
    )
    
    print("✅ ParameterManager created successfully!")
    print(f"   Reference PTA: {param_manager.reference_pta}")
    print(f"   Combine components: {param_manager.combine_components}")
    print(f"   Add DM derivatives: {param_manager.add_dm_derivatives}")
    
    # Build parameter mappings
    print("\nBuilding parameter mappings...")
    try:
        mapping = param_manager.build_parameter_mappings()
        print("✅ Parameter mapping created successfully!")
        print(f"   Merged parameters: {len(mapping.merged_parameters)}")
        print(f"   PTA-specific parameters: {len(mapping.pta_specific_parameters)}")
        
        # Show some example parameters
        if mapping.merged_parameters:
            print(f"\nExample merged parameters: {list(mapping.merged_parameters.keys())[:5]}")
        if mapping.pta_specific_parameters:
            print(f"Example PTA-specific parameters: {list(mapping.pta_specific_parameters.keys())[:5]}")
            
    except Exception as e:
        print(f"Parameter mapping failed (expected with mock data): {e}")
        print("In a real environment with actual par files, this would work correctly")
else:
    print("Skipping ParameterManager creation - no file data available")
```
