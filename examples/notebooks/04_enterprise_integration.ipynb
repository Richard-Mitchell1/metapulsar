## 1. Prepare MetaPulsar Data

First, let's create MetaPulsar data that can be used with Enterprise.


````python
# Create file discovery service with working directory
# Note: We need to specify the working directory because the notebook runs from
# /workspaces/metapulsar/examples/notebooks/ but the data is in /workspaces/metapulsar/data/
discovery = FileDiscoveryService(working_dir="/workspaces/metapulsar")

# Discover files for specific PTAs
# Note: Using available PTAs that have data
pta_names = ["epta_dr1_v2_2", "ppta_dr2", "nanograv_9y"]
file_data = discovery.discover_all_files_in_ptas(pta_names)

print("File discovery results:")
for pta_name, files in file_data.items():
    print(f"  {pta_name}: {len(files)} file pairs")
````

````python
# Filter out PTAs with no files and group results
file_data_with_files = {pta: files for pta, files in file_data.items() if files}

factory = MetaPulsarFactory()

# Group files by pulsar
pulsar_groups = factory.group_files_by_pulsar(file_data_with_files)

print(f"Found {len(pulsar_groups)} pulsars: {list(pulsar_groups.keys())}")
````

````python
# Takes a while so only create two pulsars

# Create MetaPulsars for each pulsar
metapulsars = {}
for pulsar_name, pulsar_file_data in list(pulsar_groups.items())[:2]:
    print(f"Creating MetaPulsar for {pulsar_name}...")
    metapulsar = factory.create_metapulsar(
        pulsar_file_data,
        combination_strategy="consistent",
        reference_pta=None,
    )
    metapulsars[pulsar_name] = metapulsar
    print(f"  Created MetaPulsar: {metapulsar.name}")

print(f"Successfully created {len(metapulsars)} MetaPulsars")
````

````python

if not file_data_with_files:
    print("\nNo PTAs have files - this might happen in test environments")
    print("In a real environment, you would have actual par/tim files")
else:
    print(f"\nUsing PTAs with files: {list(file_data_with_files.keys())}")
    
    # Create MetaPulsar factory
    factory = MetaPulsarFactory()
    
    # Create MetaPulsar with consistent strategy (preferred for Enterprise)
    print("\nCreating MetaPulsar with consistent strategy...")
    reference_pta = list(file_data_with_files.keys())[0]
    
    metapulsar = factory.create_metapulsar(
        file_data_with_files,
        combination_strategy="consistent",
        reference_pta=reference_pta
    )
    
    print("✅ MetaPulsar created successfully!")
    print(f"   Strategy: {metapulsar.combination_strategy}")
    print(f"   Reference PTA: {reference_pta}")
    print(f"   Number of PTAs: {len(metapulsar.pulsars)}")
    print(f"   PTA names: {list(metapulsar.pulsars.keys())}")
    
    # Show Enterprise-ready data structure
    print("\nEnterprise-ready data structure:")
    print(f"   MetaPulsar object: {type(metapulsar).__name__}")
    print(f"   Pulsars: {len(metapulsar.pulsars)}")
    print(f"   Combination strategy: {metapulsar.combination_strategy}")
    print("   Ready for Enterprise integration: ✅")
````

## 2. Create Staggered Selection for Enterprise

Staggered selection is a key feature for Enterprise integration, allowing you to select data based on multiple criteria.


````python
# Create staggered selection function for Enterprise
print("Creating staggered selection function...")

# Create staggered selection function
selection_func = create_staggered_selection(
    name="telescope_backend_selection",
    flag_criteria={
        "telescope": None,  # Primary flag
        ("backend", "freq_band"): None,  # Fallback flags
    },
    freq_range=(50, 2000)  # Frequency range
)

print(f"✅ Staggered selection function created: {selection_func.__name__}")
print(f"   Function type: {type(selection_func).__name__}")
print(f"   Callable: {callable(selection_func)}")

# Show how to use the selection function
print("\nUsage example:")
print("   # In Enterprise, you would use this function like:")
print("   # selection = selection_func(timing_data)")
print("   # This would select data based on telescope, backend, and frequency criteria")
````
