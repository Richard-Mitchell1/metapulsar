# MetaPulsar Basic Usage Tutorial

This notebook demonstrates the basic workflow of using MetaPulsar to combine pulsar timing data from multiple PTA collaborations.

## Learning Objectives
- Understand MetaPulsar's core concepts
- Learn how to discover files from different PTAs
- Create MetaPulsar objects with different combination strategies
- Explore the basic API workflow


## 1. Import and Setup

First, let's import the necessary modules and set up our environment.


```python
# No need to modify sys.path - MetaPulsar is an installable package

from metapulsar import MetaPulsarFactory, FileDiscoveryService

print("MetaPulsar imported successfully!")
print(f"MetaPulsar version: {MetaPulsarFactory.__module__}")
```

## 2. File Discovery Service

The `FileDiscoveryService` is your entry point for discovering pulsar timing data files from different PTA collaborations.


```python
# Create file discovery service
discovery = FileDiscoveryService()

# List all available PTAs
available_ptas = discovery.list_ptas()
print(f"Available PTAs: {available_ptas}")
print(f"Total PTAs: {len(available_ptas)}")
```

## 3. Discover Files and Create MetaPulsar

Let's discover files for specific PTAs and create MetaPulsar objects with different strategies.


```python
# Discover files for specific PTAs
pta_names = ["epta_dr2", "ppta_dr2"]
file_data = discovery.discover_all_files_in_ptas(pta_names)

print("File discovery results:")
for pta_name, files in file_data.items():
    print(f"  {pta_name}: {len(files)} file pairs")
    if files:
        print(f"    Example: {files[0]['par'].name} / {files[0]['tim'].name}")

# Filter out PTAs with no files
file_data_with_files = {pta: files for pta, files in file_data.items() if files}

if not file_data_with_files:
    print("\nNo PTAs have files - this might happen in test environments")
    print("In a real environment, you would have actual par/tim files")
else:
    print(f"\nUsing PTAs with files: {list(file_data_with_files.keys())}")
    print(f"Total file pairs: {sum(len(files) for files in file_data_with_files.values())}")
    
    # Create MetaPulsar factory
    factory = MetaPulsarFactory()
    print("\nMetaPulsarFactory created successfully!")
    
    # Create MetaPulsar with composite strategy
    print("\nCreating MetaPulsar with composite strategy...")
    metapulsar_composite = factory.create_metapulsar(
        file_data_with_files,
        combination_strategy="composite"
    )
    
    print("✅ MetaPulsar created successfully!")
    print(f"   Strategy: {metapulsar_composite.combination_strategy}")
    print(f"   Number of PTAs: {len(metapulsar_composite.pulsars)}")
    print(f"   PTA names: {list(metapulsar_composite.pulsars.keys())}")
    
    # Create MetaPulsar with consistent strategy
    print("\nCreating MetaPulsar with consistent strategy...")
    reference_pta = list(file_data_with_files.keys())[0]
    
    metapulsar_consistent = factory.create_metapulsar(
        file_data_with_files,
        combination_strategy="consistent",
        reference_pta=reference_pta
    )
    
    print("✅ MetaPulsar created successfully!")
    print(f"   Strategy: {metapulsar_consistent.combination_strategy}")
    print(f"   Reference PTA: {reference_pta}")
    print(f"   Number of PTAs: {len(metapulsar_consistent.pulsars)}")
```
