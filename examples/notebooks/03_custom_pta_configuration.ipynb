# MetaPulsar Custom PTA Configuration Tutorial

This notebook demonstrates how to add custom PTA configurations and use them in the MetaPulsar workflow.

## Learning Objectives
- Understand PTA configuration structure
- Learn how to add custom PTAs to the discovery service
- Explore custom PTA integration with MetaPulsar
- Handle custom PTA file patterns and settings


```python
from metapulsar import FileDiscoveryService, MetaPulsarFactory

print("MetaPulsar Custom PTA Configuration imported successfully!")
```

## 1. Understanding PTA Configuration

PTA configurations define how to discover and process files for a specific PTA collaboration.


```python
# Create file discovery service
discovery = FileDiscoveryService()

# List existing PTAs
existing_ptas = discovery.list_ptas()
print(f"Existing PTAs: {existing_ptas}")

# Show example configuration structure
print("\nExample PTA configuration structure:")
example_config = {
    "base_dir": "/data/custom_pta/",
    "par_pattern": r"([BJ]\d{4}[+-]\d{2,4})_custom\.par",
    "tim_pattern": r"([BJ]\d{4}[+-]\d{2,4})_custom\.tim",
    "timing_package": "pint",
    "priority": 1,
    "description": "Custom PTA for demonstration"
}

for key, value in example_config.items():
    print(f"  {key}: {value}")

print("\nConfiguration fields:")
print("  base_dir: Directory where PTA files are stored")
print("  par_pattern: Regex pattern for par files")
print("  tim_pattern: Regex pattern for tim files")
print("  timing_package: Timing package to use (pint, tempo2)")
print("  priority: Priority for file selection (higher = more important)")
print("  description: Human-readable description of the PTA")
```

## 2. Add Custom PTA Configuration

Let's add a custom PTA configuration to the discovery service.


```python
# Add custom PTA configuration
custom_pta_name = "custom_pta"
custom_config = {
    "base_dir": "/data/custom_pta/",
    "par_pattern": r"([BJ]\d{4}[+-]\d{2,4})_custom\.par",
    "tim_pattern": r"([BJ]\d{4}[+-]\d{2,4})_custom\.tim",
    "timing_package": "pint",
    "priority": 1,
    "description": "Custom PTA for demonstration"
}

print(f"Adding custom PTA: {custom_pta_name}")
discovery.add_pta(custom_pta_name, custom_config)
print("✅ Custom PTA added successfully!")

# List all PTAs (including custom one)
all_ptas = discovery.list_ptas()
print(f"\nAll PTAs (including custom): {all_ptas}")
print(f"Total PTAs: {len(all_ptas)}")

# Verify custom PTA was added
if custom_pta_name in all_ptas:
    print(f"✅ Custom PTA '{custom_pta_name}' is now available!")
else:
    print(f"❌ Custom PTA '{custom_pta_name}' was not added properly")
```

## 3. Use Custom PTA with MetaPulsar

Now let's use the custom PTA in the MetaPulsar workflow.


```python
# Discover files for custom PTA
print("Discovering files for custom PTA...")
file_data = discovery.discover_all_files_in_ptas([custom_pta_name])

print(f"File discovery results for {custom_pta_name}:")
for pta_name, files in file_data.items():
    print(f"  {pta_name}: {len(files)} file pairs")

# Filter out PTAs with no files
file_data_with_files = {pta: files for pta, files in file_data.items() if files}

if file_data_with_files:
    print(f"\nUsing custom PTA with files: {list(file_data_with_files.keys())}")
    
    # Create MetaPulsar factory
    factory = MetaPulsarFactory()
    print("\nCreating MetaPulsar with custom PTA...")
    
    # Create MetaPulsar with custom PTA
    metapulsar = factory.create_metapulsar(
        file_data_with_files,
        combination_strategy="composite"
    )
    
    print("✅ MetaPulsar created successfully with custom PTA!")
    print(f"   Strategy: {metapulsar.combination_strategy}")
    print(f"   Number of PTAs: {len(metapulsar.pulsars)}")
    print(f"   PTA names: {list(metapulsar.pulsars.keys())}")
    
else:
    print(f"\nNo files found for custom PTA '{custom_pta_name}' - this is expected in test environment")
    print("In a real environment, you would place par/tim files in the custom PTA directory")
    print("The custom PTA configuration has been successfully added to the discovery service")
    
    # Show how to use the custom PTA in a real scenario
    print("\nTo use this custom PTA in a real scenario:")
    print("1. Create the directory: /data/custom_pta/")
    print("2. Place par/tim files matching the patterns:")
    print("   - Par files: J1909-3744_custom.par")
    print("   - Tim files: J1909-3744_custom.tim")
    print("3. Run the discovery service again")
    print("4. The custom PTA will be discovered and usable with MetaPulsar")
```
